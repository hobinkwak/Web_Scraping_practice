## Selenium을 이용한 이미지 크롤링
- 유튜브 영상을 이미지로 크롤링
- 특정 영역(Element)을 자르기
    - pillow : python을 이용해서 이미지를 전처리해주는 대표적인 패키지
- 유튜브 영상 데이터를 다운로드
- fake_useragent 사용법
- headless 설정
    - 브라우저를 화면에 띄우지 않고 RAM영역에만 띄워서 사용하는 방법
        - 예전에는 팬텀제이스를 많이 사용했는데 지금은 안 씀!
    - 서버에서 selenium을 이용해서 크롤링할 때 사용
        - server에는 윈도우가 없다. 커맨드밖에 없으므로 화면에는 브라우저 못띄움.
        - 이 때 headless 사용

### 특정 영역의 element 좌표값 구하기

```python
driver = webdriver.Chrome()
driver.get(url)
```

- 특정 element의 위치값과 사이즈값 구하기

```python
element = driver.find_element_by_css_selector('#movie_player')
location = element.location
size = element.size
print(location, size)
driver.quit()
```

```python
# area 변경 : 해상도에 따라서 비율이 달라지기 때문에 비율을 맞춰 줘야 함.
left = location['x']
top = location['y']
right = left+size['width']
bottom = top + size['height']
area = (left, top, right, bottom)
area
```

### 이미지를 pillow로 잘라주기

```python
pil_element_img = pil_img.crop(area)
pil_element_img
```

### 이미저 저장

```python
pil_element_img.save('datas/screenshot_element.png')
```

### 썸네일 만들기 : 원본 이미지를 다양한 사이즈로 축소하는 방법

```python
pil_element_img.thumbnail((320, 160))
# 가로세로 비율을 유지하면서 바꾸기 때문에, 출력값 다를 수 있음
print(pil_element_img.size)
```

### Headless
- 브라우저를 화면에 띄우지 않고 메모리상에서만 실행하는 방법
- 크롬 60 버전 이상에서만 사용이 가능
- window 화면이 지원되지 않는 환경에서 사용
    - 서버

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')
driver = webdriver.Chrome(options=options)

# 패스트캠퍼스의 title 데이터를 가지고 온다.
driver.get('https://fastcampus.co.kr/')
print(driver.title)
driver.quit()
```
